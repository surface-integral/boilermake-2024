```python
question_prompts = []

for i in range (df.shape[0]):
    line = df["Question"][i]
    number0 = df["Numbers"][i].split()[0]
    number1 = df["Numbers"][i].split()[1]
    line = line[0: line.index("number0")] + str(number0) + line[line.index("number0")+7:]
    line = line[0: line.index("number1")] + str(number1) + line[line.index("number1")+7:]
    question_prompts.append("draw "+line)

df["question_prompts"] = question_prompts
df.head()
```
