```python
question_prompts = []

print(df.shape)
for i in range (df.shape[0]):
    line = df["Question"][i]
    if (len(df["Numbers"][i].split()) == 2):
        number0 = df["Numbers"][i].split()[0]
        number1 = df["Numbers"][i].split()[1]
        line = line[0: line.index("number0")] + str(number0) + line[line.index("number0")+7:]
        line = line[0: line.index("number1")] + str(number1) + line[line.index("number1")+7:]
        question_prompts.append(line)
    else: 
        df = df.drop(i)


df["question_prompts"] = question_prompts
df.head()
```

```python
import spacy 
from spacy import displacy
from spacy.lang.en.stop_words import STOP_WORDS

NER = spacy.load("en_core_web_lg")
```

```python
def contains_num(text):
    l = [True if "NUM" == token.pos_ else False for token in text]
    return (True in l)

def andrew_function(question, operation, problem_type): 
    output = {"Question":"", "Subject1":"", "Quantity1":"", "Subject2":"", "Quantity2":"", "Operation":"", "Type":""}
    output["Operation"] = operation
    output["Type"] = problem_type

    doc = NER(question)

    sents = [] 
    subject = ""
    quantity = ""
    for sent in doc.sents: 
        sents.append(sent)
    output["Question"] = sents[-1].text
    for sent in sents[:len(sents)-1]:
        for token in sent:
            if(token.pos_ == "NUM") and (output["Quantity1"] != ""):
                output["Quantity2"] = token.text
            if (token.dep_ == "nummod"): 
                subtree = list(token.subtree)
                start = subtree[0].i
                end = subtree[-1].i + 1
                j = 1
                for token2 in doc[end:]:
                    if not subject:
                        if token2.pos_ == "NOUN":
                            for token3 in doc[end:end+j]:
                                subject += token3.text+" "
                            break; 
                j += 1 
                quantity = doc[start:end].text
                if output["Quantity1"] == "":
                    #print("first quantity:" +token.text)
                    output["Quantity1"] = quantity
                else: 
                    #print("second quantity:" +token.text)
                    output["Quantity2"] = quantity

                if output["Subject1"] == "":
                    #print("first quantity:" +token.text)
                    output["Subject1"] = subject
                else: 
                    #print("second quantity:" +token.text)
                    output["Subject2"] = subject
            if (not subject) and (not quantity):
                #print(token.text+", "+token.pos_+", "+token.dep_)
                if (("subj" in token.dep_) or ("dobj" in token.dep_)):
                    subtree = list(token.subtree)
                    start = subtree[0].i
                    end = subtree[-1].i + 1
                    adj = ""
                    contains_num = [True if "NUM" == token.pos_ else False for token in doc[start:end]]
                    contains_noun = [True if "NOUN" == token.pos_ else False for token in doc[start:end]]
                    k = 0 
                    seg1 = doc[start:end]
                    seg2 = ""
                    for token in doc[start:end]:
                        if (token.pos_ == "CCONJ"): 
                            if True in [True if "NUM" == token.pos_ else False for token in doc[k+1:]]: 
                                seg1 = doc[start:end][:k]
                                seg2 = doc[start:end][k+1:]
                        k += 1 
                    for seg in (seg1, seg2): 
                        if seg: 
                            if ((True in contains_num) and (True in contains_noun)):
                                for token in seg[:contains_noun.index(True)+1]:
                                    if (token.pos_ == "NUM"): 
                                        quantity = token.text
                                        if output["Quantity1"] == "":
                                            #print("first quantity:" +token.text)
                                            output["Quantity1"] = quantity
                                        else: 
                                            #print("second quantity:" +token.text)
                                            output["Quantity2"] = quantity
                                    if ((token.pos_ == "ADJ") and (token.text != "more")): 
                                        adj = token.text
                                    if ((token.pos_ == "NOUN")):   # filter to only nouns 
                                        subject = adj+" "+token.text
                                        # Singularize token subjects
                                        # if token.tag_ == "NNS":
                                        #     subject = adj + TextBlob(token.text).words[0].singularize()
                                        if output["Subject1"] == "":
                                            #print("first quantity:" +token.text)
                                            output["Subject1"] = subject
                                        else: 
                                            #print("second quantity:" +token.text)
                                            output["Subject2"] = subject
                                        adj = ""
        if output["Subject2"] == "": 
            output["Subject2"] = subject

    return output 

output = andrew_function("linda has 34 candies . chloe has 28. how many candies do they have in all ?", "Addition", "and")
print("Question: "+output["Question"])
print("Subject1: "+output["Subject1"])
print("Quantity1: "+output["Quantity1"])
print("Subject2: "+output["Subject2"])
print("Quantity2: "+output["Quantity2"])
print("Type: "+output["Type"])
print("Operation: "+output["Operation"])
```
